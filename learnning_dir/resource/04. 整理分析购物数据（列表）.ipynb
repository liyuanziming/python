```python
print(f"帽子列表中索引为1的数据为：{hats[1]}") # 访问列表中索引为1的数据
print(f"帽子列表中索引为-1的数据为：{hats[-1]}") # 访问列表中索引为-1的数据
print(f"帽子列表中索引0到3的数据为：{hats[0:3]}") # 访问索引0到3的数据
```
