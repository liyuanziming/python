## 1.2 循环遍历获取列表中的元素

应用循环结构，可以遍历列表中的每一个值，比如，逐一打印列表中的每个元素，方法如下：

方法一：

```python
for i in range(len(hats)): # 按照列表长度来逐一遍历
    print(hats[i], end=" ") # end的作用是print()输出不换行，以空格做间隔
```

方法二：

```python
for hat in hats: # 逐个获取列表hats中的元素，并输出
    print(hat, end=" ")
```

# 3. 匹配购物数据推荐帽子（一）

## 3.1 根据顾客性别与购物数据确定推荐列表

获取顾客属性，与购物数据列表中各款帽子适合的性别进行逐一比对，属性一致的帽子，添加到待推荐列表之中。

提示：设置一个空列表，用于存储符合顾客属性的帽子。

```python
# 根据顾客性别和购物数据推荐帽子
hats_list = [] # 创建一个空列表，用于存储适合顾客的帽子编号
for i in range(len(hats)): # 循环遍历每一款帽子的购物数据，判断是否与顾客属性一致
    if genders[i] == gender : # 如果顾客性别gender与帽子的适配性别一致
        hats_list.append(hats[i]) # 将帽子添加到推荐列表
```

## 3.2 根据推荐列表完成推荐

完成本步实验，你将使用hats_recommend()函数。

提示：需要判断推荐列表是否为空。

```python
if len(hats_list) == 0: # 如果推荐列表为空
    print("没有适合你的帽子")
else: # 如果推荐列表非空
    hats_recommend(img, hats_list) # 推荐列表中的帽子
```

# 4. 匹配购物数据推荐帽子（二）

## 4.1 根据顾客性别、年龄与购物数据确定推荐列表

提示：
- 帽子适配的最小年龄存放在列表min_ages；最大年龄存放在列表max_ages中
- 如果顾客的年龄段[mini_age, max_age]的区间在帽子适配的年龄区间内，即可推荐。


### 4.1.1 提取顾客年龄段中的最小年龄与最大年龄

```python
mini_age = age[0]
max_age = age[1]
```

### 4.1.2 根据顾客性别、年龄逐项对比帽子的数据，获得推荐列表

```python
# 根据顾客年龄、性别和购物数据推荐帽子

hats_list = [] # 创建一个空列表，用于存储适合顾客的帽子编号
# 循环结构：逐项获取帽子数据
for i in range(len(hats)):
    # 外层分支：如果顾客性别gender与帽子的适配性别一致
    if genders[i] == gender: 
        # 内层分支：如果预测顾客年龄区间(mini_age,max_age)在帽子适配的最小与最大年龄区间内
        if min_ages[i] < mini_age and max_ages[i] > max_age: 
            hats_list.append(hats[i]) # 将帽子添加到推荐列表
```

## 4.2 根据推荐列表完成推荐

```python
if len(hats_list) == 0: # 如果推荐列表为空
    print("没有适合你的帽子")
else:    # 如果推荐列表非空
    hats_recommend(img, hats_list) # 推荐列表中的帽子
```
